```python
"""Graph Implementation in python"""
# what the underscores mean 
# http://igorsobreira.com/2010/09/16/difference-between-one-underline-and-two-underlines-in-python.html
# _ to be used as private (symbolic)
# __ don't overide outside its class (practical)
class Graph:
    
    def __init__(self,graph_dict=None):
        """ initializes a graph object 
            If no dictionary or None is given, 
            an empty dictionary will be used
        """
        if graph_dict is None:
            graph_dict=dict()
        self.__graph_dict=graph_dict
        
    def vertices(self):
        """ returns the vertices of a graph """
        return list(self.__graph_dict.keys())
    
    def edges(self):
        """ returns the edges of a graph """
        return self.__generate_edges()
    
    def add_vertex(self, vertex):
        """ If the vertex "vertex" is not in 
            self.__graph_dict, a key "vertex" with an empty
            list as a value is added to the dictionary. 
            Otherwise nothing has to be done. 
        """
        if vertex not in self.__graph_dict:
            self.__graph_dict[vertex] = []
            
    def add_edge(self, edge):
        """ assumes that edge is of type set, tuple or list; 
            between two vertices can be multiple edges! 
        """
        #edge = set(edge)
        (vertex1, vertex2) = tuple(edge)
        if vertex1 in self.__graph_dict:
            self.__graph_dict[vertex1].append(vertex2)
        else:
            self.__graph_dict[vertex1] = [vertex2]
            
    def __generate_edges(self):
        """ A static method generating the edges of the 
            graph "graph". Edges are represented as sets 
            with one (a loop back to the vertex) or two 
            vertices 
        """
        edges = []
        for vertex in self.__graph_dict:
            for neighbour in self.__graph_dict[vertex]:
                if {neighbour, vertex} not in edges:
                    edges.append({vertex, neighbour})
        return edges
    
    def find_path(self,start_vertex,end_vertex,path=None):
        """ Find a path between 2 given vertices"""    
        if path is None:
            path=[]
        
        if (end_vertex not in self.__graph_dict) or (start_vertex not in self.__graph_dict):
            return None
        
        path+=[start_vertex]
        
        if start_vertex == end_vertex:
            return path
        
        for vertex in self.__graph_dict[start_vertex]:
            if vertex not in path:
                extend_path = self.find_path(vertex,end_vertex,path)
                if extend_path:
                    return extend_path
        return None

    def find_all_paths(self,start_vertex,end_vertex,path=[]):
        """ Finds all paths between given vertices"""
        print("path:{}".format(path))     
        #path.extend([start_vertex])
        path=path+[start_vertex] #https://stackoverflow.com/questions/6951792/python-a-b-not-the-same-as-a-a-b
        if start_vertex == end_vertex:
            print("eq:{}".format(path))
            return [path]
             
        if (start_vertex not in self.__graph_dict):
            return []
        
        paths=[]
        
        for vertex in self.__graph_dict[start_vertex]:
            if vertex not in path:
                extend_path = self.find_all_paths(vertex,end_vertex,path)
                for p in extend_path:
                    print("extend_path p:{}".format(p))
                    paths.append(p)
                    print("paths:{}".format(paths))
        print("before return paths:{}".format(paths))
        return paths
    
    def degree(self,vertex):
        """Find degree of a vertex"""
        adj_vertices = self.__graph_dict[vertex]
        degree=len(adj_vertices) + adj_vertices.count(vertex)
        return degree
    
    def density(self):
        """Calculate density of Graph"""
        vertices=len(self.__graph_dict.keys())
        edges=len(self.edges())
        return (2*edges)/((vertices)*(vertices-1))
    
    def diameter():
        return False
    
    def __str__(self):
        res = "vertices: "
        for k in self.__graph_dict:
            res += str(k) + " "
        res += "\nedges: "
        for edge in self.__generate_edges():
            res += str(edge) + " "
        return res
```

```python
''' Shifted Binary search (get position of the product). A shelf contains some products sorted according to their width
, we place more already sorted products with lesser width than the one that are already present. |||100,200||1,22,34,45|||
find the position of a given book (no sorting allowed and find it in O(log N))
'''
# Bug fixing problem

def search_ele(shelf,lower_l,upper_l,product):
    if not shelf:
        return -1
    
    if lower_l is upper_l:
        return lower_l if shelf[lower_l] is product else -1
    
    mid = int((lower_l + upper_l)/2)
    
    if shelf[mid] is product:
        return mid
    
    # somewhere in 2nd sorted part 
    if shelf[lower_l] <= shelf[mid]:
        if product >= shelf[lower_l] and product <= shelf[mid]:
            return search_ele(shelf,lower_l,mid-1,product)
        return search_ele(shelf,mid+1,upper_l,product)
    
    if product>= shelf[mid] and product <= shelf[upper_l]:
        return search_ele(shelf,mid+1,upper_l,product)
    return search_ele(shelf,lower_l,mid-1,product)

def draw_width(width):
    print("")
    for i in range(width):
        print("-",end="")

el=[10,20,1,2]

print("########### Given Shelf ###########")
for width in reversed(el):
    draw_width(width)
print("\n###################################")

pos=search_ele(el,0,len(el)-1,10)

if pos is not -1:
    print("\n########### Found product on the Shelf ###########")
    for i,width in enumerate(reversed(el)):
        draw_width(width)
        if i is len(el)-1-pos: 
            print("found product with width:{} ".format(width),end="")
    print("\n##################################################")
else:
    print("\nNo such product!!")
    

        
```

```python
# QUICK
import random

def l_quicksort(x):
    """    Not in place sorting """
    if len(x) < 2:
        return x
    else:
        pivot = x[0]
        less = [i for i in x[1:] if i <= pivot]
        greater = [i for i in x[1:] if i > pivot]
        return l_quicksort(less) + [pivot] + l_quicksort(greater)

#################################################################

def sub_partition(array, start, end, idx_pivot):

    'returns the position where the pivot winds up'

    if not (start <= idx_pivot <= end):
        raise ValueError('idx pivot must be between start and end')

    array[start], array[idx_pivot] = array[idx_pivot], array[start]
    pivot = array[start]
    i = start + 1
    j = start + 1

    while j <= end:
        if array[j] <= pivot:
            array[j], array[i] = array[i], array[j]
            i += 1
        j += 1

    array[start], array[i - 1] = array[i - 1], array[start]
    return i - 1

def quicksort(array, start=0, end=None):

    if end is None:
        end = len(array) - 1

    if end - start < 1:
        return

    idx_pivot = random.randint(start, end)
    i = sub_partition(array, start, end, idx_pivot)
    #print array, i, idx_pivot
    quicksort(array, start, i - 1)
    quicksort(array, i + 1, end)
##############################################################

alist = [54,26,93,17,77,31,44,55,20]

quicksort(alist)
print(alist)
# print(l_quicksort(alist))

# Fibonacci
def fib(n):
    a, b = 0, 1
    while a < n:
        print(a,end=' ')
        a, b = b, a + b

fib(1000)

# elements=[10,1,7,6,2,20]    
# quick_sort(elements,int(len(elements)/2))
```

```python
# Python program for implementation of heap Sort
 
# To heapify subtree rooted at index i.
# n is size of heap
# k-ary k=2,...
# parent =(i-1)/k, children = (k*i)+1,...,(k*i)+k
def heapify(arr, n, i):
    largest = i  # Initialize largest as root
    l = 2 * i + 1     # left = 2*i + 1
    r = 2 * i + 2     # right = 2*i + 2
 
    # See if left child of root exists and is
    # greater than root
    if l < n and arr[i] < arr[l]:
        largest = l
 
    # See if right child of root exists and is
    # greater than root
    if r < n and arr[largest] < arr[r]:
        largest = r
 
    # Change root, if needed
    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]  # swap
 
        # Heapify the root.
        heapify(arr, n, largest) #heap for one entry
 
# The main function to sort an array of given size
def heapSort(arr):
    n = len(arr)
 
    # Build a maxheap.
    for i in range(n, -1, -1):
        heapify(arr, n, i)
 
    # One by one extract elements
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]   # swap
        heapify(arr, i, 0)
 
# Driver code to test above
arr = [ 1,12, 11, 13, 5, 6, 7]
heapSort(arr)
n = len(arr)
print ("Sorted array is")
for i in range(n):
    print ("%d" %arr[i],end=' ')
```

```python
# memoisation
# No. ways to climb

def ways(n,memo):
    print("n:{},memo[n]:{}".format(n, memo[n]))
    if memo[n] is not -1:
        return memo[n]
    elif n<0:
        return 0
    elif n == 0:
        return 1
    else:
        memo[n] = ways(n-1,memo) + ways(n-2,memo) + ways(n-3,memo)
        return memo[n]

n=10
memo=[-1]*(n+1)
print(memo)
ways(n,memo)
print(memo)
```
